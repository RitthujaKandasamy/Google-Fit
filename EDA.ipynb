```python
import pandas   as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv("C:\\Users\\ritth\\code\\Strive\\Google-Fit\\data\\dataset_5secondWindow.csv")
data.sample(5)
```

### Drop some columns

```python
data = data.drop(['activityrecognition#0'], axis=1)
data = data.drop(['user'], axis=1)
data = data.drop(['Unnamed: 0'], axis=1)
data = data.drop(['id'], axis=1)
```

```python
data.shape
```

### Split target and features

```python
X = data.drop(['target'], axis = 1)
Y = data['target']
```

```python
Y.value_counts()
```

### Missng values

```python
plt.figure(figsize = (30, 10))
plt.xticks(size = 20, color ='grey')
plt.tick_params(size = 12, color ='grey')
plt.title('Finding Null Values Using Heatmap\n', color ='grey', size = 30)
sns.heatmap(data.isnull(),
            yticklabels = False,
            cbar = False,
            cmap = 'PuBu_r',
            )
plt.show()
```

### Distribution of the data

```python
for i in X.columns:
    plt.figure(figsize = (10, 6))
    sns.distplot(X[i], hist_kws = dict(linewidth = 1, edgecolor = "k"), bins = 20)
    
    plt.title(i)
    plt.xlabel(i)

    
    plt.tight_layout()
    plt.show()
```

### Target

```python
# Plot the target

plt.figure(figsize = (10, 6))
#column = {'0': 'Car', '1': 'Bus', '2': 'Still', '3': 'Walking', '4': 'Train'}
sns.countplot(x='target',data= data, palette ='Set3')
plt.show()
```
